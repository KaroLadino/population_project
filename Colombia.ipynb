```python
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv ("country_population_col.csv")
```

```python
data
```

```python
predictors = ['Fertility', 'Life_Expectancy']
X = data[predictors]
y = data['Population']

lm = LinearRegression()
model = lm.fit(X, y)
```

```python
print(f'alpha = {model.intercept_}')
print(f'betas = {model.coef_}')
```

```python
new_population = model.predict(X)
```

```python
data ['new_population'] = new_population.tolist()
```

```python
data
```

```python
pred_2018 = [[1.853, 74.381]]
print(model.predict(pred_2018))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Anio'], data['Population'], '*')           # scatter plot showing actual data
plt.plot(data['Anio'], data['new_population'], '*')
#plt.plot(data['Population'], model, 'r', linewidth=2)   # regression line
plt.xlabel('Population')
plt.ylabel('Anio')
plt.title('Expected Population Vs. new population')
plt.legend()
plt.show()
```
