```python
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv ("C:/Users/Ladin/country_population_col.csv")
```

```python
data
```

```python
# Build linear regression model using TV and Radio as predictors
# Split data into predictors X and output Y
predictors = ['Fertility', 'Life_Expectancy']
X = data[predictors]
y = data['Population']

# Initialise and fit model
lm = LinearRegression()
model = lm.fit(X, y)
```

```python
print(f'alpha = {model.intercept_}')
print(f'betas = {model.coef_}')
```

```python
model.predict(X)
```

```python
new_X = [[1.853, 74.381]]
print(model.predict(new_X))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Population'], data['Life_Expectancy'], 'o')           # scatter plot showing actual data
#plt.plot(data['Population'], model, 'r', linewidth=2)   # regression line
plt.xlabel('Population')
plt.ylabel('Life Expectancy')
plt.title('Population vs Life Expectancy')

plt.show()
```
